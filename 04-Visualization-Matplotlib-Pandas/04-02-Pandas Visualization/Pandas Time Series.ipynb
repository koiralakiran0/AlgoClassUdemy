```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as dates
%matplotlib notebook
```

```python
mcdon = pd.read_csv('mcdonalds.csv', index_col='Date', parse_dates=True)
```

```python
mcdon.head()
```

```python
mcdon['Adj. Close'].plot()
```

```python
mcdon['Adj. Volume'].plot(figsize=(12,4))
```

```python
mcdon['Adj. Close'].plot(xlim=['2007-01-01', '2009-01-01'], ylim = (20,50))
```

```python
idx = mcdon.index
```

```python
idx = mcdon.loc['2007-01-01':'2007-05-01'].index
```

```python
idx
```

```python
stock = mcdon.loc['2007-01-01':'2007-05-01']['Adj. Close']
```

```python
stock
```

```python
fig, ax = plt.subplots()
ax.plot_date(idx, stock, '-')

#Locating
ax.xaxis.set_major_locator(dates.MonthLocator())
#Formatting
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n\n\n%b %Y'))

ax.xaxis.set_minor_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_minor_formatter(dates.DateFormatter('%a'))

fig.autofmt_xdate()
fig.tight_layout()
```
